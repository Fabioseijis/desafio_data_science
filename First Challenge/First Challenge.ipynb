```python
gastos_anuais = gastos_anuais.set_index('Unidade da Federação')
```

```python
gastos_anuais_2019 = gastos_anuais[2019]
```

```python
gastos_anuais_2019.sort_values(ascending = False).plot(kind='bar')
```

```python
gastos2019_percentagem = gastos2019 / gastos2019.sum() * 100
```

```python
gastos2019_percentagem
```

```python
gastos_anuais_2019_percentagem = gastos_anuais_2019 / gastos_anuais_2019.sum() * 100
```

```python
gastos_anuais_2019_percentagem.sort_index()
```

```python
gastos2019_percentagem.sort_index()
```

```python
comparacao2019 = pd.concat([gastos_anuais_2019_percentagem, gastos2019_percentagem], axis=1)
comparacao2019
```

```python
comparacao2019.columns = ['Produção Hospitalar 2019', 'Produção Hospitalar per capta 2019']
```

```python
comparacao2019
```

```python
comparacao2019.plot(kind = 'bar', figsize=(15,8), grid = True)
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
plt.title('Produção hospitalar e produção hospitalar per capta dos estados brasileiros em relação ao total')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

##### Apesar do estado de São Paulo ter a maior produção hospitalar do Brasil, quando vemos em relação per capta dessa produção podemos observar que São Paulo fica um pouco acima da média nacional, justificando a alta produção. Por outro lado, os estados da região sul, principalmente o Paraná detem os maiores valores de produção per capta do país. Uma rápida pesquisa revela que uma das possíveis causas de uma produção mais elevadas nesses estados, é o maior salário dos médicos nas redes públicas de saúde. Do outro lado do espectro vemos que o contrário acontece no Amapá, onde a população não é muito expressiva e a produção hospitalar é a menor do país quase que 5x menor que o do Paraná.
